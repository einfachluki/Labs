### Cleaning and wrangling data

- Some car brand names refer to the same brand. Replace all brand names that contain the word "Dutton" for simply "Dutton". If you find similar examples, clean their names too. Use `loc` with boolean indexing.

- Convert CO2 Emissions from Grams/Mile to Grams/Km

- Create a binary column that solely indicates if the transmission of a car is automatic or manual. Use `pandas.Series.str.startswith` and .

- convert MPG columns to km_per_liter

```python
#1
df.loc[df['Make'].str.contains('Dutton'), 'Make'] = df['Make'].apply(lambda x:'Dutton')
df.loc[df['Make'].str.contains('Dutton')]
```

Note:
<br>Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Converting Gallons to Liters

1 Gallon = 3.78541 Liters



```python
##2
km = 1.60934

df["CO2 Emission Grams / KM"] = df["CO2 Emission Grams/Mile"]/km
column1 = df.pop("CO2 Emission Grams / KM")
df.insert(df.columns.get_loc("CO2 Emission Grams/Mile"), "CO2 Emission Grams / KM", column1)
df
```

```python
df.Transmission.value_counts()
```

```python
##3
pd.Series.str.startswith
df["AUTOMATIC?"] = df["Transmission"].str.startswith("Auto")
df
```

```python
#convert MPG columns to km_per_liter

df["City km_liter"]= df["City MPG"]*(1.60934/3.78541)
df["Highway km_liter"]= df["Highway MPG"]*(1.60934/3.78541)
df
```
